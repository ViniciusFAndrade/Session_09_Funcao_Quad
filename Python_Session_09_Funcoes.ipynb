```
#Função quadrática
def quad(a,b,c):
    y1 = - b + ((b**2-4*a*c)**0.5)/(2*a)
    y2 = - b - ((b**2-4*a*c)**0.5)/(2*a)
    return(y1,y2)
quad(1,20,3)
```
